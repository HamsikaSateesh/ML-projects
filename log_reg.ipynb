```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Cheking for NULL Values

```python
df.isnull().sum()
```

### treat null/missing values


```python
mcols = ['MARRIAGE','PAY_SEPT','PAY_AUG', 'PAY_JUL','PAY_JUN', 'PAY_MAY', 'PAY_APR']

for col in mcols:
    #mod = df[col].mode
    df.fillna(value={col : df[col].mode()[0]}, inplace=True)
```

```python
df.isnull().sum()
```

### Checking for duplicate values

```python
df.duplicated().sum()
```

```python
df.drop(columns='ID',inplace=True)
```

```python
df.duplicated().sum()
```

```python
#df = df.drop_duplicates()
df = df.drop_duplicates(keep='first')
```

```python
df.duplicated().sum()
```

```python
df.shape
```

### Converting Categorical Data into Numeric Data for model understanding
#### 1. One Hot Encoding & Dummies creation - Nominal Data
#### 2. Label Encoding - Ordinal Data
#### 3. Binay Encoding - T/F values

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['EDUCATION'])
```

```python
le.classes_
```

```python
df['EDUCATION'] = le.transform(df['EDUCATION'])
```

```python
mapping = dict(zip(le.classes_, range(len(le.classes_))))
print("Mapping of values to categories:")
for category, encoded_value in mapping.items():
    print(f"{category}: {encoded_value}")
```

```python
#Label encoding- One hot encoding
encode_num = {"SEX":{"Female":0,"Male":1}, "IsDefaulter":{"Yes":1,"No":0}, "EDUCATION":{"University":3,"Graduate School":2,"High School":1,"Others":0}}
data=df.replace(encode_num)
```

```python
df.head()
```

```python
#data = pd.get_dummies(data, columns = [''])
data = pd.get_dummies(data)
```

```python
data.head()
```

### Check for data balance in the dependent variable

```python
data['IsDefaulter'].value_counts()
```

```python
x = data.loc[:,data.columns != 'IsDefaulter']
y = data['IsDefaulter']
```

### Different ways to balance the data
##### 1. Using SMOTE (Synthetic Minority Over-sampling Technique)
##### 2. Using ADSYN ADASYN (Adaptive Synthetic Sampling)

```python
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
```

```python
xts , yts = smote.fit_resample(x,y)
```

```python
print("Values after balancing :",len(yts))
print("Values before balancing :",len(y))
```

### Transforming the data - Feature Engineering
#### 1. Standardization
#### 2. Normalization

```python
#importing libraries for data transformation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xts = scaler.fit_transform(xts)
```

#### Splitting the Data into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xts, yts, test_size=0.2, random_state=42)
```

```python
print("Samples in Training data ", len(X_train))
print("Samples in Testing data ", len(X_test))
```

#### Model Training / Data Modelling & Model Evaluation

```python
#importing models and evaluation metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
```

#### Logistic Regression

```python
#logistic regression
logi = LogisticRegression(fit_intercept=True, max_iter=10000)
logi.fit(X_train, y_train)
y_pred_logi = logi.predict(X_test)
y_train_pred_logi=logi.predict(X_train)

#getting all score for logistic regression
train_accuracy_logi = round(accuracy_score(y_train_pred_logi,y_train), 3)
accuracy_logi = round(accuracy_score(y_pred_logi,y_test), 3)
precision_score_logi = round(precision_score(y_pred_logi,y_test), 3)
recall_score_logi = round(recall_score(y_pred_logi,y_test), 3)
f1_score_logi = round(f1_score(y_pred_logi,y_test), 3)
roc_score_logi = round(roc_auc_score(y_pred_logi,y_test), 3)

print("The accuracy on train data is ", train_accuracy_logi)
print("The accuracy on test data is ", accuracy_logi)
print("The precision on test data is ", precision_score_logi)
print("The recall on test data is ", recall_score_logi)
print("The f1 on test data is ", f1_score_logi)
print("The roc_score on test data is ", roc_score_logi)

labels = ['Not Defaulter', 'Defaulter']
cm_logi = confusion_matrix(y_test, y_pred_logi )
print(cm_logi)
```
